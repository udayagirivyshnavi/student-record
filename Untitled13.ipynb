```python
students = []

while True:
    sid = int(input("Enter Student ID: "))
    name = input("Enter Name: ")
    course = input("Enter Course: ")

    students.append({
        "id": sid,
        "name": name,
        "course": course
    })

    choice = input("Add another student? (yes/no): ").lower()

    if choice == "no":
        break

print("\nStudent Records:")
for student in students:
    print(student)
```
